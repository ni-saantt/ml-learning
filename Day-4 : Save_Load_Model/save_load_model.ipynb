```python
import pandas as pd
import numpy as np
from sklearn import linear_model

df = pd.read_csv("/content/sample_data/homeprices (3).csv")
df.head()

model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)

model.coef_
model.intercept_

model.predict([[5000]])

#saving model by using pickle

import pickle

with open('model_pickle.pkl','wb') as file:
    pickle.dump(model,file)

#loading it

with open("model_pickle.pkl", "rb") as file:
    mp = pickle.load(file)
mp.coef_
mp.intercept_
mp.predict([[5000]])

#saving model using joblib
#it is recomended to use

import joblib

joblib.dump(model, 'model_joblib')

#loading it

mj = joblib.load('model_joblib')
mj.coef_
mj.intercept_
mj.predict([[5000]])
```
