```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df=pd.read_csv("/content/sample_data/city_profit.csv")
df
%matplotlib inline
plt.xlabel("population")
plt.ylabel("profit")
plt.scatter(df.population,df.profit,color='red',marker='*')

reg=linear_model.LinearRegression()
reg.fit(df[['population']],df.profit)
reg.predict([[27]])

g=pd.read_csv("/content/sample_data/population_test.csv")
g
g['predicted_profit']= reg.predict(g)
g

x_line=df[['population']]
y_line=reg.predict(x_line)
plt.plot(x_line, y_line)

%matplotlib inline
plt.figure(figsize=(8,6))

#orri
plt.scatter(df.population, df.profit, color='red', marker='*', label='Actual Data')



#line
plt.plot(df.population, y_line, color='blue', label='Regression Line')

#new points output
plt.scatter(g.population, g.predicted_profit, color='green', marker='o', label='Predicted Data')

plt.xlabel("Population ( 10,000 people)")
plt.ylabel("Profit ( $10,000)")

plt.title("City Population vs Profit ")


plt.legend()

plt.show()


```
